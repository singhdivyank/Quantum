```python
%matplotlib inline
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile
from qiskit.providers.aer import QasmSimulator
from qiskit.visualization import plot_histogram # to plot a histogram
```

```python
circuit = QuantumCircuit(2, 2) # Quantum circuit

# draw the circuit
circuit.draw()
```

```python
circuit.h(0) # puts 0 into superposition state
circuit.cx(0, 1)
circuit.measure([0, 1], [0, 1])
```

```python
circuit.draw() # entanglement
```

```python
simulator = QasmSimulator() # simulator to be used
compiled_circuit = transpile(circuit, simulator)
job = simulator.run(compiled_circuit, shots=1000)
result = job.result()
counts = result.get_counts(circuit)
print("\nTotal count for 00 and 11 are:",counts)
```

```python
plot_histogram(counts)
```
